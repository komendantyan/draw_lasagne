```python
import cPickle
# from collections import OrderedDict

import theano
import theano.tensor as T
import lasagne

import utils
import advanced_layers
```

```python
mnist = utils.load_mnist()
binary_mnist = (
    ((mnist[0][0] > 0.8).astype('float32'), mnist[0][1]),
    ((mnist[1][0] > 0.8).astype('float32'), mnist[1][1]),
)
```

```python
input_layer = lasagne.layers.InputLayer((32, 1, 28, 28), name='input')
conv_1_layer = lasagne.layers.Conv2DLayer(input_layer, 50, (3, 3), pad='same', name='conv_1')
conv_2_layer = lasagne.layers.Conv2DLayer(conv_1_layer, 50, (3, 3), stride=(2, 2), pad='same', name='conv_2')
conv_3_layer = lasagne.layers.Conv2DLayer(conv_2_layer, 50, (3, 3), pad='same', name='conv_3')
conv_4_layer = lasagne.layers.Conv2DLayer(conv_3_layer, 50, (3, 3), stride=(2, 2), pad='same', name='conv_4')
conv_5_layer = lasagne.layers.Conv2DLayer(conv_4_layer, 50, (3, 3), pad='same', name='conv_5')

dense_1_layer = lasagne.layers.DenseLayer(conv_5_layer, 100, name='dense_1')

mean_layer = lasagne.layers.DenseLayer(dense_1_layer, 5, name='mean', nonlinearity=None)
logvar2_layer = lasagne.layers.DenseLayer(dense_1_layer, 5, name='logvar2', nonlinearity=None)
sample_layer = advanced_layers.SamplingLayer([mean_layer, logvar2_layer], name='sample')

dense_2_layer = lasagne.layers.DenseLayer(sample_layer, 100, name='dense_2')
dense_3_layer = lasagne.layers.DenseLayer(dense_2_layer, 50*7*7, name='dense_3')
reshape_1_layer = lasagne.layers.ReshapeLayer(dense_3_layer, (32, 50, 7, 7), name='reshape_1')

deconv_1_layer = lasagne.layers.Deconv2DLayer(reshape_1_layer, 50, (3, 3), crop='same', name='deconv_1')
deconv_2_layer = lasagne.layers.Deconv2DLayer(deconv_1_layer, 50, (3, 3), stride=(2, 2), crop='same',
                                              output_size=(14, 14), name='deconv_2')
deconv_3_layer = lasagne.layers.Deconv2DLayer(deconv_2_layer, 50, (3, 3), crop='same', name='deconv_3')
deconv_4_layer = lasagne.layers.Deconv2DLayer(deconv_3_layer, 50, (3, 3), stride=(2, 2), crop='same',
                                              output_size=(28, 28), name='deconv_4')
deconv_5_layer = lasagne.layers.Deconv2DLayer(deconv_4_layer, 1, (3, 3), crop='same', 
                                              nonlinearity=lasagne.nonlinearities.sigmoid, name='deconv_5')
```

```python
for layer in lasagne.layers.get_all_layers(deconv_5_layer):
    print "{name}\t{output_size}".format(name=layer.name, output_size=layer.output_shape[1:])
```

```python
mean, logvar2, encoded, decoded = lasagne.layers.get_output([mean_layer, logvar2_layer, sample_layer, deconv_5_layer])
image = input_layer.input_var

bc_loss = T.mean(lasagne.objectives.binary_crossentropy(decoded, image))
kl_loss = 0.5 * T.mean(mean**2 + T.exp(logvar2) - logvar2 - 1)
loss = (bc_loss + 0.01 * kl_loss)

updates = lasagne.updates.adadelta(loss, lasagne.layers.get_all_params(deconv_5_layer), learning_rate=1.0)
```

```python
train_batch = theano.function([image], loss, updates=updates)
test_batch = theano.function([image], loss)
predict_batch = theano.function([image], decoded)

encode_batch = theano.function([image], encoded)
decode_batch = theano.function([encoded], decoded)
```

```python
train = utils.batch_wrapper(train_batch, 32, aggregate=numpy.mean)
test = utils.batch_wrapper(test_batch, 32, aggregate=numpy.mean)
predict = utils.batch_wrapper(predict_batch, 32, cut=True, shuffle=False)

encode = utils.batch_wrapper(encode_batch, 32, cut=True, shuffle=False)
decode = utils.batch_wrapper(decode_batch, 32, cut=True, shuffle=False)
```

```python
weights_value = lasagne.layers.get_all_param_values(deconv_5_layer)
```

```python
lasagne.layers.set_all_param_values(deconv_5_layer, weights_value)
```

```python
NB_EPOCH = 100
BATCH_SIZE = 32

train_loss = [float('inf')] * 4
test_loss = [float('inf')] * 4
for epoch in xrange(NB_EPOCH):
    print 'epoch', epoch, '\t',
    train_loss.append(train(mnist[0][0]))
    print "train_loss", train_loss[-1], '\t',
    test_loss.append(test(mnist[1][0]))
    print "test_loss", test_loss[-1]
    if min(test_loss[-3:]) > test_loss[-4]:
        print 'early stopping'
        break
```
