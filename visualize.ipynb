```python
%pylab inline
import utils
import glob

from IPython.display import HTML, display
```

```python
mnist = utils.load_mnist(process=lambda x: (x > 0.8).astype('float32'))
```

```python
experiments = sorted(glob.glob('predictions/*'))

for experiment in experiments:
    data_by_steps = [load(step) for step in sorted(glob.glob(experiment + '/step*.npy'))]
    if not data_by_steps:
        continue
    images = [utils.collage(data[:64], (8, 8)) for data in data_by_steps]
    gif_path = experiment + '/anim.gif'
    utils.make_gif(images, gif_path)
    
    display(HTML(r'%s<br><img src="%s">' % (gif_path, gif_path)))
```
